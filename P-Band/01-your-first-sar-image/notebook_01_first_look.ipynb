---

## Step 2: Load Data


```python
import os
import rasterio

# Define the path to the sample data
filepath = '../data/sample_sar_image.tif'

# Safety check to ensure the data exists
if not os.path.exists(filepath):
    print(f"⚠️ Error: Could not find {filepath}. Did you run the download script in Step 1?")
else:
    # Open the raster file
    with rasterio.open(filepath) as src:
        # Read the first band (index 1 in rasterio)
        sar_band = src.read(1)
        # Extract the metadata profile
        sar_profile = src.profile
        
    print("✅ Data loaded successfully!")
    print(f"Image dimensions (Pixels): {sar_band.shape}")
    print(f"Coordinate Reference System (CRS): {sar_profile['crs']}")
```

---

## Step 3: Process and Visualize


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Clean the data: Avoid log(0) errors by setting zero or negative values to NaN
sar_band_clean = np.where(sar_band > 0, sar_band, np.nan)

# 2. Convert raw linear intensity to Decibels (dB)
sar_db = 10 * np.log10(sar_band_clean)

# 3. Calculate percentiles to stretch the contrast (ignoring NaNs)
vmin = np.nanpercentile(sar_db, 2)
vmax = np.nanpercentile(sar_db, 98)

# 4. Create the visualization
plt.figure(figsize=(10, 8))

# Display the image using a standard grayscale colormap for SAR
im = plt.imshow(sar_db, cmap='gray', vmin=vmin, vmax=vmax)

# Add presentation elements
plt.colorbar(im, label='Backscatter Intensity (dB)')
plt.title('My First SAR Image Visualization', fontsize=14, pad=15)
plt.xlabel('Pixel Columns')
plt.ylabel('Pixel Rows')

# 5. Save the output as a high-resolution PNG
output_filename = 'my_first_sar_image.png'
plt.savefig(output_filename, dpi=300, bbox_inches='tight')

# Render the plot in the notebook
plt.show()

print(f"🎉 Mission Complete! Visualization saved as '{output_filename}'.")
```
