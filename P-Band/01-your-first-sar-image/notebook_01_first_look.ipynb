## Step 2: Load Data


```python
import os
import rasterio

# Define the path to the sample data
filepath = '../data/sample_sar_image.tif'

# Safety check to ensure the data exists
if not os.path.exists(filepath):
    print(f"⚠️ Error: Could not find {filepath}. Did you run the download script in Step 1?")
else:
    # Open the raster file
    with rasterio.open(filepath) as src:
        # Read the first band (index 1 in rasterio)
        sar_band = src.read(1)
        # Extract the metadata profile
        sar_profile = src.profile
        
    print("✅ Data loaded successfully!")
    print(f"Image dimensions (Pixels): {sar_band.shape}")
    print(f"Coordinate Reference System (CRS): {sar_profile['crs']}")
```
